# Primitives possibles

```python
import numpy as np
import matplotlib.pyplot as plt


size = 4000

inX_cercle  = []
inY_cercle  = []
outX_cercle = []
outY_cercle = []

inX_ring  = []
inY_ring  = []
outX_ring = []
outY_ring = []

inX_rectT  = []
inY_rectT  = []
outX_rectT = []
outY_rectT = []

inX_T  = []
inY_T  = []
outX_T = []
outY_T = []

inX_8  = []
inY_8  = []
outX_8 = []
outY_8 = []

inX_star  = []
inY_star  = []
outX_star = []
outY_star = []

inX_U  = []
inY_U  = []
outX_U = []
outY_U = []

# cercle
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if( ((x-0.5)**2  + (y-0.5)**2) < .25):
        inX_cercle.append(x)
        inY_cercle.append(y)
    else:
        outX_cercle.append(x)
        outY_cercle.append(y)

# # ring
for i in range(size):

    x = np.random.rand()
    y = np.random.rand()

    if( .015625 < ((x-0.5)**2  + (y-0.5)**2) < .25):
        inX_ring.append(x)
        inY_ring.append(y)
    else:
        outX_ring.append(x)
        outY_ring.append(y)


# rectangular triangle
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if (x+y < 1):
        inX_rectT.append(x)
        inY_rectT.append(y)
    else:
        outX_rectT.append(x)
        outY_rectT.append(y)

# triangle
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if (((x < 0.5) and (2*x > y))or((x>0.5) and (y<-2*x+2))):
        inX_T.append(x)
        inY_T.append(y)
    else:
        outX_T.append(x)
        outY_T.append(y)


# 8 faces
for i in range(size):
    x     = np.random.rand()
    y     = np.random.rand()

    if ((x+y < 5./3) and (x+y > 1./3) and (y > x-2./3) and (y<x+2./3)):
        inX_8.append(x)
        inY_8.append(y)
    else:
        outX_8.append(x)
        outY_8.append(y)

# Star
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    lowerTriangle = ((x < 0.5) and (-8./5*x+0.8< y < 0.8))or((x>0.5) and (8./5*x-0.8<y<0.8))
    upperTriangle = ((x < 0.5) and (0.2< y < 8./5*x+0.2))or((x > 0.5) and (0.2< y < -8./5*x+9./5))

    if (lowerTriangle or upperTriangle):
        inX_star.append(x)
        inY_star.append(y)
    else:
        outX_star.append(x)
        outY_star.append(y)

# U-shape
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if(((y<.5) and ( .015625 < ((x-0.5)**2  + (y-0.5)**2) < .25)) or ((y>.5)  and ((0.375 > x) or (x > 0.625))) ):
        inX_U.append(x)
        inY_U.append(y)
    else:
        outX_U.append(x)
        outY_U.append(y)


plt.figure(figsize=(13, 6))
plt.suptitle("possible 2D-Primitive", fontsize=16)

plt.subplot(241)
plt.plot(inX_cercle,inY_cercle,'.',c='g')
plt.plot(outX_cercle,outY_cercle,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.subplot(242)
plt.plot(inX_ring,inY_ring,'.',c='g')
plt.plot(outX_ring,outY_ring,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.subplot(243)
plt.plot(inX_rectT,inY_rectT,'.',c='g')
plt.plot(outX_rectT,outY_rectT,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.subplot(244)
plt.plot(inX_T,inY_T,'.',c='g')
plt.plot(outX_T,outY_T,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.subplot(245)
plt.plot(inX_8,inY_8,'.',c='g')
plt.plot(outX_8,outY_8,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.subplot(246)
plt.plot(inX_star,inY_star,'.',c='g')
plt.plot(outX_star,outY_star,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.subplot(247)
plt.plot(inX_U,inY_U,'.',c='g')
plt.plot(outX_U,outY_U,'.',c='r')
plt.xlim((0,1))
plt.ylim((0,1))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


size = 20000

inX_sphere  = []
inY_sphere  = []
inZ_sphere  = []
inX_cercle  = []
inY_cercle  = []


# 2D
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if( ((x-0.5)**2  + (y-0.5)**2) < .25):
        inX_cercle.append(x)
        inY_cercle.append(y)
        
# 3D
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()
    z = np.random.rand()

    if( ((x-0.5)**2  + (y-0.5)**2) + (z-0.5)**2 < .25):
        inX_sphere.append(x)
        inY_sphere.append(y)
        inZ_sphere.append(z)


fig = plt.figure(figsize=(10, 10))
plt.suptitle("Sphere", fontsize=16)
ax  = fig.add_subplot(111, projection='3d')
ax.autoscale(enable=True, axis='both', tight=None)

ax.scatter(np.array(inX_sphere), inY_sphere, inZ_sphere, c='g', marker='.', zdir='z')
plt.show()
fig2D = plt.figure(figsize=(10, 10))
plt.plot(inX_cercle,inY_cercle,'.',c='g')
plt.xlim((0,1))
plt.ylim((0,1))
plt.show()
```

```python
size = 20000

inX_tor  = []
inY_tor  = []
inZ_tor  = []
inX_cercle = []
inY_cercle = []

a = 0.5
b = 0.2


# 2D
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if( (a-b)**2< ((x-0.5)**2  + (y-0.5)**2) < a**2):
        inX_cercle.append(x)
        inY_cercle.append(y)
        
# 3D
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()
    z = np.random.rand()

    if( (((x-.5)**2+(y-.5)**2+(z-.5)**2-(a**2+b**2))**2 - 4*a**2*(b**2-(z-.5)**2)) < 0):
        inX_tor.append(x)
        inY_tor.append(y)
        inZ_tor.append(z)


fig = plt.figure(figsize=(10, 10))
plt.suptitle("Torus", fontsize=16)
ax  = fig.add_subplot(111, projection='3d')
ax.scatter(inX_tor, inY_tor, inZ_tor, c='g', marker='.', zdir='z')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,1)
plt.show()


fig2D = plt.figure(figsize=(10, 10))
plt.plot(inX_cercle,inY_cercle,'.',c='g')
plt.xlim((0,1))
plt.ylim((0,1))
plt.show()


```

```python
size = 20000

inX_tri3D = []
inY_tri3D = []
inZ_tri3D = []
inX_tri = []
inY_tri = []


# 2D
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()

    if((.5 - (.2) < x < .5 + (.2))&(.5 - (.2) < y < .5 + (.2))):
        inX_tri.append(x)
        inY_tri.append(y)
        
# 3D
for i in range(size):
    x = np.random.rand()
    y = np.random.rand()
    z = np.random.rand()

    if((.5 - (1-z)/2 < x < .5 + (1-z)/2)&(.5 - (1-z)/2 < y < .5 + (1-z)/2)):
        inX_tri3D.append(x)
        inY_tri3D.append(y)
        inZ_tri3D.append(z)


fig = plt.figure(figsize=(10, 10))
plt.suptitle("pyramid", fontsize=16)
ax  = fig.add_subplot(111, projection='3d')
ax.scatter(inX_tri3D, inY_tri3D, inZ_tri3D, c='g', marker='.', zdir='z')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,1)
plt.show()


fig2D = plt.figure(figsize=(10, 10))
plt.plot(inX_tri,inY_tri,'.',c='g')
plt.xlim((0,1))
plt.ylim((0,1))
plt.show()
```
